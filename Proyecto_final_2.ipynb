```python
from scipy import stats
import numpy as np
import pandas as pd 
import math
import matplotlib.pyplot as plt
from random import seed
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.preprocessing import LabelBinarizer
```

Aquí se definen las operaciones de la tabla de verdad. La operación $x_{1}\&x_{2}$ se define como $x_{1}yx_{2}$ en el código, mientras que la operación $x_{1}\$x_{2}$ se define como $x_{1}px_{2}$ en el código. 

Por otra parte, se toma un arreglo de (4,2) para describir los valores que toman $x_{1}$ y $x_{2}$ en la tabla de verdad, por lo cual, las operaciones \$,\& son funciones que toman arreglos de (4,2) y regresan un arreglo de (4,1). 

Por último, dado que A toma valores distintos de 1 y 0 en algunos casos, estos se redondean para definir las operaciones binarias del problema en cuestión.

```python
AyB=np.array([(1.001,0),(0,1),(0.001,0), (1,1)]).round()
x1ux2=np.array([(0,0),(1,0),(0,1),(1,1)])
x1yx2 = np.array([1,1,1,0])
x1px2 = np.array([0,0,0,1])
```

Se propone para las dos operaciones dos redes neuronales que contienen cada una 2 entradas, 4 neuronas y 1 salida. 

Las primeras 4 neuronas se activan por la función 'relu', mientras que la de salida se activa por la función 'sigmoidal'. Ésta última la definimos así, ya que, una tabla de verdad (en esencia) es una clasificación binaria.

Para el entrenamiento de las redes neuronales se propone como optimizador el método 'adam', como función de perdida 'mean_squared_error' y como métrica 'binary_accuracy', ésta última tomada así porque se busca una clasificación binaria; sin embargo, para el entrenamiento de la primera red neuronal se hacen 700 iteraciones mientras que en la segunda se hacen 500 iteraciones. 

```python
model1 = Sequential()
model1.add(Dense(units=4, activation='relu',input_dim=2))
model1.add(Dense(units=1, activation='sigmoid'))
model1.compile(optimizer='adam', loss='mean_squared_error', metrics=['binary_accuracy'])
model1.fit(x1ux2,x1yx2,epochs=700)
```

Como se puede observar, se obtiene una exactitud del 100\% para la operación $\&$

```python
_,accuracy=model1.evaluate(x1ux2,x1yx2)
print('Accuracy:%.2f'% (accuracy*100))
```

```python
model1.predict(x1ux2).round()
```

```python
model2 = Sequential()
model2.add(Dense(units=4, activation='relu',input_dim=2))
model2.add(Dense(units=1, activation='sigmoid'))
model2.compile(optimizer='adam', loss='mean_squared_error', metrics=['binary_accuracy'])
model2.fit(x1ux2,x1px2,epochs=500)
```

Y también se obtiene una exactitud del 100\% para la operación \$

```python
_,accuracy=model2.evaluate(x1ux2,x1px2)
print('Accuracy:%.2f'% (accuracy*100))
```

```python
model2.predict(x1ux2).round()
```
